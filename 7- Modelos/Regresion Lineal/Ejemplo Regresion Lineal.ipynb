Machine learning 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
casas.head()
```

```python
casas.columns
```

Division Datos de Pruebas y Datos de Entrenamiento

```python
X = casas[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
Y = casas['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

Entrenamiento del modelo

```python
lrm = LinearRegression()
lrm.fit(X_train, y_train)
```

Creacion del modelo de pruebas - Evaluacion del modelo de entrenamiento

```python
from sklearn import metrics
```

```python
predicciones = lrm.predict(X_test)
predicciones
```

```python
y_test
```

```python
plt.scatter(y_test, predicciones)
```

```python
sns.set_context('notebook', font_scale=0.7)
sns.distplot(y_test - predicciones)
```

Metricas

MAE(Mean absolute error) - Media del valor absoluto de los errores

```python
metrics.mean_absolute_error(y_test, predicciones)
```

MSE (Media de los errores al cuadrado)

```python
metrics.mean_squared_error(y_test, predicciones)
```

RMSE 

```python
np.sqrt(metrics.mean_squared_error(y_test, predicciones))
```

```python
print('Training score {:.2f}'.format(lrm.score(X_train, y_train)))
print('Test score {:.2f}'.format(lrm.score(X_test, y_test)))
```
