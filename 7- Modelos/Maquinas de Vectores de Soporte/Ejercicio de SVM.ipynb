MAQUINAS DE VECTORES DE SOPORTE - CLASIFICACION Y REGRESION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flores = sns.load_dataset('iris')
```

```python
flores.head()
```

```python
flores['species'].unique()
```

```python
sns.pairplot(flores, hue='species')
```

DIVISION DE DATOS

```python
X = flores.drop('species', axis=1)
y = flores['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

ENTRENAMIENTO DEL MODELO

```python
from sklearn.svm import SVC
```

```python
modelo = SVC(gamma='auto')
modelo.fit(X_train, y_train)
```

```python
predicciones = modelo.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicciones))
```

```python
print(confusion_matrix(y_test, predicciones))
```

```python
print('Training score {:.2f}'.format(modelo.score(X_train, y_train)))
print('Test score {:.2f}'.format(modelo.score(X_test, y_test)))
```
