```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y,'red')
plt.title("Este es mi grafico")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

```python
plt.subplot(1,2,1) # Una fila, dos columnas, numero 1
plt.plot(x,y,'g')

plt.subplot(1,2,2) #Una fila , dos columnas, numero 2
plt.plot(y,x,'blue')
```

Programacion orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'green')
```

```python
figura = plt.figure()
grafico1= figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2= figura.add_axes([0.15,0.5,0.4,0.4])
grafico2.plot(y,x,'blue')
```

Multigraficos(filas, columnas)

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)
graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'red')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'green')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarillo')
graficos[1][1].set_title('Verde')
```

Tamao del Grafico

```python
figura = plt.figure(figsize=(10,2)) #Tamao en pulgadas
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

Crear dos graficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label='Al cuadrado')
graficos.plot(x,x**3,label='Al cubo')
graficos.legend(loc=0) #Agrega leyenda
```

Color del grafico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='black')
```

Linea del grafico(grosor,transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'r',linewidth=10,alpha=0.5,linestyle=':') #Cordenadas, color, grosor, transparencia, estilo
```
