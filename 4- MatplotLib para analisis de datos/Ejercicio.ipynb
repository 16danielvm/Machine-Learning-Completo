Ejercicio 


Crear una variable "x" que contenga 10 valores equidistantes entre los valores 0 y 20
Crear una variable "y" que contenga valores de la funcion "x al cuadrado +5"
Crear un grafico color rojo, con ancho de 5 y estilo de puntos "..."

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,10)
x
```

```python
y = (x**2 + 5)
y
```

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red',linewidth=5, linestyle=':')
```
